```python
Task3 = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
      STATE, COUNT(CLAIM_AMOUNT) AS NUMBER_OF_CLAIMS, RISK_SEGMENTATION AS RISK
FROM  
      insurance_data
WHERE
      RISK = 'H'
GROUP BY  
      STATE
ORDER BY 
      NUMBER_OF_CLAIMS DESC
LIMIT 5;      
   

---------------------WRITE QUERY ABOVE---------------------
""")

Task3
```
