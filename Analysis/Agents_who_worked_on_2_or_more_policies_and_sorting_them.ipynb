```python
employee_data = pd.read_csv('employee_data.csv')
employee_data
```

```python
Task7 = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
       COUNT(DISTINCT INSURANCE_TYPE) AS TYPES_OF_POLICY, E.AGENT_NAME, SUM(CLAIM_AMOUNT) AS TOTAL_CLAIM_AMOUNT
FROM 
       insurance_data I JOIN employee_data E ON I.AGENT_ID = E.AGENT_ID
GROUP BY 
       E.AGENT_NAME
HAVING 
       COUNT(INSURANCE_TYPE) > 2
ORDER BY 
       TOTAL_CLAIM_AMOUNT DESC;       
           
   

---------------------WRITE QUERY ABOVE---------------------
""")

Task7
```
