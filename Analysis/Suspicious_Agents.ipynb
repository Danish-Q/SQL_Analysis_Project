```python
insurance_data = pd.read_csv('insurance_data.csv')
insurance_data
```

```python
Task12 = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
       E.AGENT_NAME, E.AGENT_ID, I.AGENT_ID, I.CLAIM_AMOUNT,
       CASE WHEN CLAIM_AMOUNT >= 15000 THEN 'SUSPISIOUS'
       ELSE 'NOT'
       END AS 'SUSPICIOUS'
FROM   
       insurance_data I JOIN employee_data E ON I.AGENT_ID = E.AGENT_ID;                  
   

---------------------WRITE QUERY ABOVE---------------------
""")

Task12
```

```python
Task12A = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
       CLAIM_STATUS, RISK_SEGMENTATION, INCIDENT_SEVERITY,
       CASE WHEN CLAIM_STATUS = 'A' AND RISK_SEGMENTATION = 'H' AND INCIDENT_SEVERITY = 'Major Loss' THEN 1
       ELSE 0
       END AS 'SUSPICIOUS'
FROM   
       insurance_data I JOIN employee_data E ON I.AGENT_ID = E.AGENT_ID;                  
   

---------------------WRITE QUERY ABOVE---------------------
""")

Task12A
```

```python
mean = Task12A.mean()
mean
```
