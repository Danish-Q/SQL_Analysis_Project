```python
Task11 = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
       E.AGENT_NAME, E.AGENT_ID, SUM(I.CLAIM_AMOUNT) AS TOTAL_CLAIM            
FROM   
       insurance_data I JOIN employee_data E ON I.AGENT_ID = E.AGENT_ID
GROUP BY 
       E.AGENT_NAME
ORDER BY 
       SUM(CLAIM_AMOUNT)
LIMIT
       5;       


          
   

---------------------WRITE QUERY ABOVE---------------------
""")

Task11
```
