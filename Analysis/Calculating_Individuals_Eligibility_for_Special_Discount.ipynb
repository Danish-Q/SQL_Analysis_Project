```python
Task9 = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
       TENURE, EMPLOYMENT_STATUS, NO_OF_FAMILY_MEMBERS,
CASE   WHEN  TENURE > 60 AND EMPLOYMENT_STATUS = 'N' AND NO_OF_FAMILY_MEMBERS >= 4 THEN 1
       ELSE 0
       END AS ELIGIBLE_FOR_DISCOUNT              
FROM   insurance_data;
          
   

---------------------WRITE QUERY ABOVE---------------------
""")

Task9
```

```python
mean = Task9.mean()
mean
```
